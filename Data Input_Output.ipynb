Pandas는 데이터 파일을 읽어 데이터프레임을 만들 수 있다. 다음처럼 여러가지 포맷을 지원한다. (csv, excel, html,json,sql 등)

```python
%%writefile sample1.csv
c1, c2, c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
import pandas as pd
```

```python
pd.read_csv('sample1.csv')
```

```python
%%writefile sample2.csv
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
pd.read_csv('sample2.csv', names=['c1', 'c2', 'c3'])
```

```python
pd.read_csv('sample1.csv', index_col='c1')  #특정한 열을 행 인덱스로 지정
```

```python
%%writefile sample3.txt
c1        c2        c3        c4
0.179181 -1.538472  1.347553  0.43381
1.024209  0.087307 -1.281997  0.49265
0.417899 -2.002308  0.255245 -1.10515
```

```python
#데이터를 구분하는 구분자가 쉼표가 아닌경우
#sep 인수를 써서 구분자를 사용자가 지정해준다.
#구분자 길이가 정해지지 않은 공백인 경우 \s+문자열을 쓴다

pd.read_table('sample3.txt', sep='\s+')
```

```python
#자료 파일 중에 건너 뛰어야 할 행이 있으면 skiprows 인수를 사용한다.

%%writefile sample4.txt
c1, c2, c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
pd.read_csv('sample4.txt', skiprows=[0, 1])
```

```python
%%writefile sample5.csv
c1, c2, c3
1, 1.11, one
2, , two
누락, 3.33, three
```

```python
df = pd.read_csv('sample5.csv', na_values=['누락'])
df
```

```python
df.to_csv('sample6.csv')
```
