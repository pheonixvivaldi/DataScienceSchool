```python
df3 = pd.DataFrame({
    'A': [1, 3, 4, 3, 4],
    'B': [2, 3, 1, 2, 3],
    'C': [1, 5, 2, 4, 4]
})
df3
```

```python
df3.apply(lambda x: x.max() - x.min())

#람다 함수 : 열의 최대값과 최소값의 차이
```

```python
df3.apply(lambda x: x.max() - x.min(), axis=1)   #행에 적용
```

```python
df3.apply(pd.value_counts)

#각 열에 대해 어떤 값이 얼마나 사용되었는지 알고 싶다면 value_counts 함수를 넣을 수도 있다.
```

```python
df3.apply(pd.value_counts).fillna(0).astype(int)

#NaN 값은 fillna 메서드를 사용하여 원하는 값으로 바꿀 수 있다. 
#astype 메서드로 전체 데이터의 자료형을 바꾸는 것도 가능하다.
```

실수 값을 카테고리 값으로 변환
1) cut : 실수 값의 경계선을 지정하는 경우
2) qcut : 갯수가 똑같은 구간으로 나누는 경우

```python
ages = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 100]
```

```python
bins = [1, 15, 25, 35, 60, 99]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
cats = pd.cut(ages, bins, labels=labels)
cats
```

```python
type(cats)
```

```python
cats.categories
```

```python
cats.codes
```

```python
df4 = pd.DataFrame(ages, columns=["ages"])
df4["age_cat"] = pd.cut(df4.ages, bins, labels=labels)
df4
```

```python
data = np.random.randn(1000)
cats = pd.qcut(data, 4, labels=["Q1", "Q2", "Q3", "Q4"])
cats
```

```python
pd.value_counts(cats)
```
